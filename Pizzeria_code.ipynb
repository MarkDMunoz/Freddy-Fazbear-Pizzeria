```python
# setting up our workstation and loading the data into pandas data frame
import pandas as pd
import matplotlib as mpt
import numpy as np
import datetime as dt
```

```python
pd.read_csv("Pizza_Results.csv")
df = pd.read_csv("Pizza_Results.csv")
```

```python
df
```

```python
# Dropping the columns that were duplicated
df.drop(['pizza_type_id.1', 'pizza_id.1','order_id.1'], axis = 1, inplace=True)
```

```python
df
```

```python
# checking for null values, there is no values to replace in categorical or numerical columns
df.isna().sum()
```

```python
# printing data type columns
df.dtypes
```

```python
# Date and time are not being displayed as date and are seen as objects
df['date'] = pd.to_datetime(df['date'])
df['time'] = pd.to_datetime(df['time'])
```

```python
print(df['date'])
print(df['time'])
```

```python
# i want to see how much pizza was ordered during the amount of time in business 
pizzas_per_day = df.groupby(df['date'].dt.date)['quantity'].sum()
```

```python
pizzas_per_day
```

```python
pizzas_per_day.plot()
```

```python
pizza_counts = df.groupby('pizza_id')['quantity'].sum()

# Sort the pizza counts in descending order
sorted_pizza_counts = pizza_counts.sort_values(ascending=False)
```

```python
print(sorted_pizza_counts.head(10))
```

```python
# Calculating Revenue 
revenue = df['price'] * df['quantity']

print(f'the business is generating {revenue.sum()} this year')
```
